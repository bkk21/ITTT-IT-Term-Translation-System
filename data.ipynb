```python
import pandas as pd

# 파일 읽기
with open('100.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

data = {'instruction':[], 'input': [], 'output': []}

lines[0]
```

```python
#lines 분리 및 정리
for i in range(len(lines)):
    if '^' in lines[i]:
        data['instruction'].append('영어 문장을 한국어로 번역하세요.')
    elif '&' in lines[i]:
        data['input'].append(lines[i].replace('&', '').strip())
    elif '*' in lines[i]:
        data['output'].append(lines[i].replace('*', '').strip())

result_data = pd.DataFrame(data)
result_data
```

```python
#데이터 백업
result_data.to_json('data.json', orient='records', indent = 4)
```
